I have used similar way to fill null values for “Fireplace Qu”. I have used where function and np.median to fill the null values in Lot Frontage. I have repeated the same thing with “Bsmt Cond”, “Bsmt Exposure”, “BsmtFinType1”, “BsmtFinType2”, “Electrical”.
